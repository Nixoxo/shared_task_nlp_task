```python
import xml.etree.ElementTree as ET
```

```python
test_file_csv = "textProcessing_testKaldi.csv"
```

```python
def create_prompt_response_map(gram):
    prompt_response_map = {}
    for prompt_unit in gram:
        key = None
        responses = []
        for prompt in prompt_unit:
            if prompt.tag == None:
                print("None")
            else:
                if(prompt.tag == 'prompt'):
                    key = prompt.text
                elif(prompt.tag == 'response'):
                    responses.append(prompt.text)
        prompt_response_map[key] = responses
    return prompt_response_map

def read_grammar_and_create_map(file):
    tree = ET.parse(file)
    grammar = tree.getroot()
    return create_prompt_response_map(grammar)
    
prompt_response_map = read_grammar_and_create_map('referenceGrammar.xml')
#prompt_response_map_3 = read_grammar_and_create_map("man_rg.xml")
```

```python
test_data = None
with open(test_file_csv, "r") as reader:
    test_data = reader.readlines()
```

```python
print("Es sind insgesamt %s prompts" % str(len(test_data[1:])))
```

```python
def count_prompts():
    count_prompt_map = {}
    for item in test_data[1:]:
        split = item.split("\t")
        prompt = split[1]
        if prompt in count_prompt_map:
            count_prompt_map[prompt] += 1
        else:
            count_prompt_map[prompt] = 1
            
    return count_prompt_map
            
count_prompt_map = count_prompts()
```

```python
def id_transcript():
    id_transcript_map = {}
    for item in test_data[1:]:
        split = item.split("\t")
        id_ = split[0]
        prompt = split[1]
        rec_result = split[3].replace("\n", "")
        id_transcript_map[id_] = {"transcript": rec_result, "prompt": prompt}
    return id_transcript_map

id_transcript_map = id_transcript()
id_transcript_map
```

```python
kaldi_data = None
with open("decoding/1_log/decode.1.log", "r") as reader:
    kaldi_data = reader.readlines()
```

```python
kaldi_prompt_sent = {}
for item in kaldi_data:
    replaced = item.replace(" \n", "")
    kaldi_prompt_sent[str(replaced[:4])] = replaced[5:]
```

```python
new_id_transcript_kaldi_map = {}
sames = 0
with open("first_kaldi_result.csv", "w") as writer:
    for key in id_transcript_map:
        #transcript = id_tran
        #print(key)
        transcript = id_transcript_map[key]['transcript'].strip()
        prompt = id_transcript_map[key]['prompt'].strip()
        
        if "***" in transcript:
            #print("passed")
            continue
        kaldi = kaldi_prompt_sent[key].lower()
        same = False
        if transcript == kaldi:
            same = True
            sames +=1
        writer.write(key + "\t" + prompt + "\t" + transcript + "\t" + kaldi + "\t"+ str(same)+"\n")
    
print(sames)   
#new_id_transcript_kaldi_map
```
